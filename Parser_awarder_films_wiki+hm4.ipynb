```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Academy_Award_for_Best_Picture'
```

```python
response=requests.get(url)
```

```python
response.content
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(response.text)
```

```python
len(soup.findAll('table'))
```

```python
tables = soup.find_all('table',{'class':'wikitable sortable'})
```

```python
import re
```

```python
Year = []
Film = []
Studio = []
for table in tables:
  for row in table.findAll('tr'):
    try:
      if len(row.find_all('td'))==1:
        year = re.split("\(|\n|/",row.find_all('td')[0].get_text() )[0]
      elif len(row.find_all('td'))==2:
        Year.append(year)
        Film.append(row.find_all('td')[0].get_text().split('\n')[0])
        if(len(row.find_all('td')[1].get_text().split('\n')[0])>1):
          print(re.split(",",row.find_all('td')[1].get_text().split('\n')[0]))
        Studio.append(row.find_all('td')[1].get_text().split('\n')[0])
    except Exception as error:
        print(error)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = {'Year': Year, 'Film': Film, 'Studio': Studio})
```

```python
df.tail(10)
```

```python
stroka = ""
```

```python
for studio in Studio:
  stroka+=studio
  stroka+=" "
```

```python
stroka
```

```python
print("    ")
print(len(re.findall(r"Steven Spielberg",stroka)))
```

```python
difference = len(df[df['Studio']=='Warner Bros.'])-len(df[df['Studio']=='Paramount'])
if difference>=0:
  print("Warner bros. ",difference)
else:
  print('Paramount ',difference)
```
